```python
from dataclasses import dataclass, field
from math import sqrt


@dataclass
class Ponto:
    x: float
    y: float
    z: float


@dataclass
class Linha:
    orig: Ponto
    dest: Ponto
    tamanho: float = field(init=False)

    def __post_init__(self):
        self.tamanho = sqrt((self.dest.x - self.orig.x)**2 + (self.dest.y - self.orig.y)**2 + (self.dest.z - self.orig.z)**2)


A = Ponto(0, 0, 0)
B = Ponto(0, 10, 0)
C = Ponto(10, 10, 0)
D = Ponto(10, 0, 0)
ponto_querido = Ponto(0, 0, -15)

pontos_fixos = A, B, C, D
```

```python
from matplotlib import pyplot as plt


fig = plt.figure(dpi=150)
ax = fig.add_subplot(111, projection='3d')

ax.scatter(ponto_querido.x, ponto_querido.y, ponto_querido.z)

for ponto in pontos_fixos:
    linha = Linha(orig=ponto_querido, dest=ponto)
    ax.scatter(ponto.x, ponto.y, ponto.z)
    ax.plot((ponto_querido.x, ponto.x),
            (ponto_querido.y, ponto.y), 
            (ponto_querido.z, ponto.z),
            label=linha)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.legend(loc='center right', bbox_to_anchor=(3.8, 0))

plt.show()
```
